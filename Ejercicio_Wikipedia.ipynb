El siguiente trozo de código permite realizar una petición HTTP a la api de Wikipedia para extraer la primera sección de texto perteneciente a la página buscada (campo *titles*). En este caso, se emplea para extraer la descripción que aparece en la siguiente página: [https://en.wikipedia.org/wiki/Aristotle](https://en.wikipedia.org/wiki/Aristotle)

```python
import requests

S = requests.Session()

URL = "https://en.wikipedia.org/w/api.php"

#row[1] contiene el id del articulo en wikipedia. row[2] el nombre del personaje histórico
PARAMS = {
    "action": "query",
    "format": "json",
    "titles": "Aristotle",
    "prop": "extracts",
    "explaintext": "1",
    "exintro": "1"
}
try:
    
    R = S.get(url=URL, params=PARAMS)
    DATA = R.json()

    print(type(DATA))
    print(DATA)
    

except Exception as e:
  print("Ha ocurrido un error "+str(e.__class__))
```

```python
DATA['query']['pages'][list(DATA['query']['pages'].keys())[0]]['extract']
```
